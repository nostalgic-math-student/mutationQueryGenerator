```python
from collections import defaultdict

# Assuming each area now includes 'Rows' and 'indexLabel' for demonstration
## TO DO: Change format of Rows from :[[301,304],[301,306]] ---> {length: [301,306], numRows: 2}
areas = [
    {
        "Reversed": True,
        "id": "01",
        "title": "General Admission",
        "name": "General Admission",
        "total":1000,
    },
]
    # {
    #     "Reversed": True,
    #     "id": "01",
    #     "title": "General Plus Admission",
    #     "name": "General Plus Admission",
    #     "total":1300,
    # },
    # {
    #     "Reversed": True,
    #     "id": "01",
    #     "title": "General VIP Admission",
    #     "name": "General VIP Admission",
    #     "total":1000,
    # },
    # {
    #     "Reversed": True,
    #     "id": "01",
    #     "title": "Deluxe Admission",
    #     "name": "Deluxe Admission",
    #     "total":1000,
    # },


prices_and_fees = {
    "General Admission": {"price": "19.99", "fees": "0.00"},
    "General Plus Admission": {"price": "29.99", "fees": "0.00"},
    "General VIP Admission": {"price": "39.99", "fees": "0.00"},
    "Deluxe Admission": {"price": "49.99", "fees": "0.00"},
}
```

```python
grouped_areas = defaultdict(list)
for area in areas:
    grouped_areas[area["name"]].append(area)
```

```python
for zone_name, areas in grouped_areas.items():
    print(zone_name)
    for area in areas:
        print(area['total'])
        print(range(area['total']))
 
```

```python
def generate_seat_data(area, seat):
    seat_str = "Ticket for "
    seat_str = seat_str +  area + " " + "no. " + str(seat) 
    return seat_str 
```

```python
title = "Cinco de Mayo - La Original Banda El Limón de Salvador Lizárraga"
description = "Cinco de Mayo show featuring La Original Banda el Limón de Salvador Lizárraga at Clayton's South Padre Island. "

def generate_gql_mutation_with_seats(grouped_areas, prices_and_fees):
    mutation = ["mutation CreateEvent {"]
    mutation.append("  createEvent(")
    mutation.append("    data: {")
    mutation.append('      description: "Soundtrack de la Naranja Mecanica en vivo.",')
    mutation.append('      name: "La Naranja Mecanica",')
    mutation.append("      eventZones: {")
    mutation.append("        create: [")

    for zone_name, areas in grouped_areas.items():
        mutation.append(
            f'          {{ name: "{zone_name}", fees: "{prices_and_fees[zone_name]["fees"]}", price: "{prices_and_fees[zone_name]["price"]}", eventZoneRows: {{ create: ['
        )
        for area in areas:
            seats_str = ""
            
            for seat_index in range(1,area['total'] + 1):
                seat_data = generate_seat_data(zone_name, seat_index)
                seats_str = seats_str +", " + '{{ name: "{0}", subZoneTag: "{1}" }}'.format(seat_data, seat_index) 
            print("seat data", seats_str)
            mutation.append(
                f'                {{ name: "{area["title"]}", eventSeats: {{ create: [{seats_str}] }} }},'
                )
        mutation.append("              ] } },")

    mutation.append("        ]")
    mutation.append("      }")
    mutation.append("    }")
    mutation.append("  ) {")
    mutation.append("    id")
    mutation.append("  }")
    mutation.append("}")

    return "\n".join(mutation)

# Generate the mutation string with variable prices, fees, dynamically generated seat names, and subZoneTags
corrected_mutation_str_with_subzone = generate_gql_mutation_with_seats(
    grouped_areas, prices_and_fees
)
```

```python
corrected_file_path = "Notebook_corrected_gql_mutation.txt"
with open(corrected_file_path, "w") as file:
    file.write(corrected_mutation_str_with_subzone)

```
